課題2：【モダンデータモデリング（メダリオン構造）】（Snowflake / dbt）
〜Upworkで最も需要が高く、最高単価を叩き出せるMDSの構築〜

ビジネス背景：
クライアント企業が1枚の巨大な生テーブル（KaggleのCSV）をそのまま運用しており、
データの更新コストが高く、どれが信頼できるデータかわからない状態になっている。
これをモダンデータスタック（MDS）の標準である「メダリオン構造」に再構築してほしい。

タスク：
dbtを使って、以下の3つのレイヤーに美しくモデリングを分解せよ。

Staging（stg_credit_card.sql）：
生データのデータ型を適正化（金額は NUMBER(18,2)、頻度は FLOAT など）し、
課題1で行った初期クレンジングを施すレイヤー。

Intermediate（int_customer_behavior_metrics.sql）：
顧客の行動指標を計算するレイヤー。前述の「全支払回数に対する分割払いの割合のパーセンタイル」や、
「キャッシング依存度（CASH_ADVANCE / PURCHASES）」を、Window関数を用いて全会員に対して計算・付与する。

Marts（fct_credit_risk_mart.sql）：
マーケティングチームやリスク管理チームがそのままBI（Tableau等）に繋いで使える最終成果物。
限度額に対する利用比率や、全額支払い割合（PRC_FULL_PAYMENT）を元に、顧客を High_Risk / Normal / Royal などのセグメントに分類して格納する。

Intermediate（中級）として意識すべき裏側の動き：
dbtの source や ref を正しく使い、データリネージ（依存関係のグラフ）を美しく保てるか。
dbt test（unique, not_null）を仕込み、データ品質が保証されている状態を自動で作れるか。
